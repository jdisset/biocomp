## Simple Hill functions for gene expression
Equations for mRNA and Protein are basically identical. Protein can also encode for an ERN, in which case Aff is not equal to 0. The mRNA can also include an ERN recognition site - again, in which case Aff is not equal to 0.

### ODE form
__mRNA__
$$ \frac{d[mRNA]}{dt} = V_{DNA} (\frac{[DNA]}{[DNA] + K_{A_{DNA}}})^{n_{DNA}} - \mu_{mRNA} [mRNA] - Aff [mRNA] [ERN] $$
__Protein__
$$ \frac{d[P]}{dt} = V_{mRNA} (\frac{[mRNA]}{[mRNA] + K_{A_{mRNA}}})^{n_{mRNA}} - \mu_{P} [P] - Aff [mRNA] [ERN] $$

### Steady state (regular)
__mRNA__
$$ [mRNA]^* = V_{DNA} (\frac{[DNA]^*}{[DNA]^* + K_{A_{DNA}}})^{n_{DNA}} / (\mu_{mRNA} + Aff [ERN]^*) $$
__Protein__
$$ [P]^* = V_{mRNA} (\frac{[mRNA]^*}{[mRNA]^* + K_{A_{mRNA}}})^{n_{mRNA}} / (\mu_{P} + Aff [mRNA]^*) $$

### Steady state (inverted)
__DNA from mRNA SS__
$$ [DNA]^* = \frac{\alpha K_{A_{DNA}}}{1 - \alpha} $$
where $\alpha$ is:
$$ \alpha = \sqrt[n_{DNA}]{\frac{[mRNA]^*}{V_{DNA}} (\mu_{mRNA}+Aff[ERN]^*)} $$
__mRNA from Protein SS__
$$ [mRNA]^* = \frac{\alpha K_{A_{mRNA}}}{1 - \alpha} $$
where $\alpha$ is:
$$ \alpha = \sqrt[n_{mRNA}]{\frac{[P]^*}{V_{mRNA}} (\mu_{P}+Aff[mRNA]^*)} $$
